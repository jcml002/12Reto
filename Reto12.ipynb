```python


import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from mpl_toolkits.mplot3d import Axes3D

# Simular datos
np.random.seed(3)
n = 100
distancia = np.random.uniform(1, 30, size=n)
peso = np.random.uniform(0.5, 10, size=n)
tiempo_entrega = 0.15 * distancia + 0.3 * peso + np.random.normal(0, 1, size=n)

# Crear DataFrame
df = pd.DataFrame({
    'Distancia': distancia,
    'Peso': peso,
    'Tiempo': tiempo_entrega
})

# Graficar
fig = plt.figure(figsize=(6, 4))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(df['Distancia'], df['Peso'], df['Tiempo'], c='teal', marker='o')
ax.set_xlabel('Distancia (km)')
ax.set_ylabel('Peso (kg)')
ax.set_zlabel('Tiempo de entrega (d칤as)')
ax.set_title('Relaci칩n entre Distancia, Peso y Tiempo de Entrega')
plt.tight_layout()
plt.show()



```
